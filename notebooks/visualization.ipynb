#### Visualization the data for better understanding


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance_cleaned.csv")
```

#### Univariate Analysis
#### Fraud Distribution

```python
sns.countplot(x='fraud_reported', data=df)
```

#### Incident Severity

```python
sns.countplot(x='incident_severity', data=df)
```

#### Claim Amount Distribution

##### claimed amount is in Indian Rupees

```python
sns.histplot(df['total_claim_amount'])
```

#### Bivariate Analysis
#### Severity vs Fraud

```python
sns.countplot(
    x='incident_severity',
    hue='fraud_reported',
    data=df
)
```

#### Incident Type vs Fraud

```python
sns.countplot(
    x='incident_type',
    hue='fraud_reported',
    data=df
)
```

#### Authorities Contacted vs Fraud

```python
sns.countplot(
    x='authorities_contacted',
    hue='fraud_reported',
    data=df
)
```

#### Correlation Heatmap

```python
corr = df.select_dtypes(exclude='object').corr()

sns.heatmap(corr)
```

####  Interpretation :

#### Higher injury/property/vehicle claims lead to higher total claim amounts.
