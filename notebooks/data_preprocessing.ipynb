#### converting dates

```python
df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'])
df['incident_date'] = pd.to_datetime(df['incident_date'])
```

#### doing Feature Engineering 


```python
df['claim_delay_days'] = (
    df['incident_date'] -
    df['policy_bind_date']
).dt.days
```

#### calculating the age of the vechile

```python
df['vehicle_age'] = 2025 - df['auto_year']
```

#### removing unneccessary columns for better performance

```python
df.nunique().sort_values(ascending=False)
```

```python
drop_cols = [
    'policy_number',
    'insured_zip',
    'incident_location'
]

df.drop(columns=drop_cols, inplace=True)
```
